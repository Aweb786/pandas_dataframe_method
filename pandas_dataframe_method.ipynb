# apply() on dataframe

```python
data=pd.DataFrame({
    'point_A':[(2,3),(5,6),(7,8),(4,8),(9,6)],
    'point_B':[(4,5),(7,8),(6,2),(5,7),(6,9)]
})
```

```python
data
```

```python
def eculidian(row):
    pt_a=row['point_A']
    pt_b=row['point_B']
    return ((pt_a[0] - pt_b[0])**2 + (pt_a[1] - pt_b[1])**2)
             

```

```python
distance=data.apply(eculidian,axis=1)
```

```python
data['distance']=data.apply(eculidian,axis=1)
```
